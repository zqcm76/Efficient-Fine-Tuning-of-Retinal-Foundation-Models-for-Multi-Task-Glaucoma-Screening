```python
import os
import sys
import torch
import pandas as pd
from torch.utils.data import DataLoader

# 固定路径，适配Notebook
CUR_DIR = r"E:\RETFound\phase1"
sys.path.insert(0, CUR_DIR)

from config import Config
from datasets import REFUGE2Dataset
from models import ResNet34Classifier
from transforms import build_cls_transforms
from engine import evaluate_classifier
from utils import get_device

# 1. 加载配置
cfg = Config("baseline.yaml")
device = get_device()

# 2. 读取数据集，仅筛选验证集
df_all = pd.read_csv(os.path.join(CUR_DIR, "data/data_all.csv"))
val_df = df_all[df_all["split"] == "val"].reset_index(drop=True)
val_trans = build_cls_transforms(train=False, size=224)
val_ds = REFUGE2Dataset(val_df, transforms=val_trans, task="cls")
val_loader = DataLoader(val_ds, batch_size=8, shuffle=False)

# 3. 加载最优权重
model = ResNet34Classifier(num_classes=1).to(device)
weight_path = r"E:\RETFound\phase1\outputs\baseline_resnet34_best.pt"
ckpt = torch.load(weight_path, map_location=device)
model.load_state_dict(ckpt["model"])
model.eval()

# 4. 计算验证集指标
metrics = evaluate_classifier(model, val_loader, device)
print("===== 验证集评估指标 =====")
for name, value in metrics.items():
    print(f"{name}: {value:.4f}")
```

```python
import os
import sys
import torch
from PIL import Image

# 固定路径，适配Notebook
CUR_DIR = r"E:\RETFound\phase1"
sys.path.insert(0, CUR_DIR)

from config import Config
from models import ResNet34Classifier
from transforms import build_cls_transforms
from utils import get_device

cfg = Config("baseline.yaml")
device = get_device()
model = ResNet34Classifier(num_classes=1).to(device)
ckpt = torch.load(r"E:\RETFound\phase1\outputs\baseline_resnet34_best.pt", map_location=device)
model.load_state_dict(ckpt["model"])
model.eval()

import numpy as np
from PIL import Image

# 预处理
trans = build_cls_transforms(train=False, size=224)
img = Image.open(r"E:\RETFound\REFUGE2\Val\img\1201.jpg").convert("RGB")
# PIL转numpy数组
img_np = np.array(img)
# Albumentations传入ndarray
aug_result = trans(image=img_np)
img_tensor = aug_result["image"].unsqueeze(0).to(device)

# 推理
with torch.no_grad():
    logit = model(img_tensor)
    prob = torch.sigmoid(logit).item()

print(f"青光眼患病概率: {prob:.3f}")
print("预测：青光眼" if prob>0.5 else "预测：正常眼底")
```
